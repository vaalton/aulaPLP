**Texto escolhido: Resolução do problema do Titanic do Kaggle**

Encontrei essa resolução do problema do titanic e achei muito pertinente para a matéria que estamos estudando, já resolvi esse problema a algum tempo atrás, foi bom rever esse problema e entender quais são as palavras mais utilizadas, até mesmo em um texto de auxílio a resolução de um problema

```python
import pandas as pd                                                                                                 #importando biblioteca para manipulação de data frames
import re                                                                                                           #importando biblioteca para leitura de arquivos
from collections import Counter                                                                                     #importando biblioteca e função para criação do dicionário e contagem das palavras
words = re.findall(r'\w+', open('Titanic Kaggle - Resolução.txt').read().lower())                                   #procurar o arquivo dentro do diretório e baixo-lo para a variável 'words" 
```


Montar um dicionário com a palavra e a quantidade de vezes que a palavra aparece no texto.

```python
counter = Counter(words)                                                                                            #função usada para geração do dicionário e contagem das palavras
print(counter)                                                                                                      #printar o dicionário com as keys (palavras) e os values (valores)
```

```python
df = pd.DataFrame(counter.items(),  columns=['word', 'count']).sort_values('count',ascending=False)                #Criação de um dataframe a partir do dicionário criado
df = df.sort_values('count',ascending=False).reset_index(drop=True)                                                #Organização do dataframe para mostrar as palavras mais repetidas
df.head()
```

```python
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

df['log_count'] = np.log(df['count'])
df['log_rank'] = np.log(df.index)
df['log_rank'].replace(-np.Inf,0,inplace=True)
plt.figure(figsize=(24,10))
sns.lineplot(x=df[df['count'] > 10].word,y=df[df['count'] > 10]['log_count'])
plt.xticks(rotation=90,fontsize='small') 

# Importando bibliotecas para geração de gráficos e manipulação de números, criação de 2 novas colunas (log baseado na frequência da palavras e ranking desse valor)
# Além de plotar o gráfico, reduzindo para análise de apenas as palavras que se repetem mais de 10 vezes

plt.show()
```

```python
contagem = sum(counter.values())                #contagem de palavras dentro do arquivo
print('O dicionário possui',contagem,'palavras')
```

**Conclusão**

Como podemos ver, muitas preposições aparecem no texto (normal em qualquer texto), para um análise mais específica como uma nuvem de palavras, o correto seria retirar essas termos para entendermos realmente quais são as palavras que aparecem mais vezes.
